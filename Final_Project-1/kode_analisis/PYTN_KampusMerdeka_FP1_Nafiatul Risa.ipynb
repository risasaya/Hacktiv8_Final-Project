## G. Modelling

### One Hot Encoding

```python
def one_hot_encoder(data, feature, keep_first=True):

    one_hot_cols = pd.get_dummies(data[feature])

    for col in one_hot_cols.columns:
        one_hot_cols.rename({col: f'{feature}_'+col}, axis=1, inplace=True)

    new_data = pd.concat([data, one_hot_cols], axis=1)
    new_data.drop(feature, axis=1, inplace=True)

    if keep_first == False:
        new_data = new_data.iloc[:, 1:]

    return new_data
```

```python
new_df_onehot = df.copy()
for col in new_df_onehot.select_dtypes(include='O').columns:
    new_df_onehot = one_hot_encoder(new_df_onehot, col)

new_df_onehot
```

### Split Dataset
Dataset di bagi 7:3, yaitu Test Data 70% train Data 30%

```python
X = new_df_onehot.drop(columns=['price'],axis=1).values
y = new_df_onehot['price'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

### Linear Regression

```python
lr = LinearRegression()
```

```python
# Fit to data training
model = lr.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print('MSE :',mse)
print('RMSE :', rmse)
```

Hasil dari Linear regression ini memiliki akurasi atau skor 93,37% dan nilai RMSE 2,26.

### Random Forest

```python
rf = RandomForestRegressor(n_estimators = 50, random_state = 13)
```

```python
rfmodel = rf.fit(X_train,y_train)
y_predrf=rfmodel.predict(X_test)
```

```python
print('R^2 Score :', r2_score(y_test, y_predrf))
```

```python
mse = mean_squared_error(y_test,y_predrf)
rmse = np.sqrt(mse)
print('MSE :',mse)
print('RMSE :', rmse)
```

Hasil dari Random Forest ini memiliki akurasi atau skor 96,50% dan nilai RMSE 1,64.
