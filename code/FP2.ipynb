# Data Cleaning

```python
#Drop Date column
df.drop("Date", axis = 1, inplace = True)
```

```python
# find categorical variables

categorical = [var for var in df.columns if df[var].dtype=='O']

print('There are {} categorical variables\n'.format(len(categorical)))

print('The categorical variables are :', categorical)
```

```python
# find numerical variables

numerical = [var for var in df.columns if df[var].dtype!='O']

print('There are {} numerical variables\n'.format(len(numerical)))

print('The numerical variables are :', numerical)
```

```python
#replacing null values in categorical variables with mode value
df['WindGustDir'].fillna(df['WindGustDir'].mode()[0], inplace=True)
df['WindDir9am'].fillna(df['WindDir9am'].mode()[0], inplace=True)
df['WindDir3pm'].fillna(df['WindDir3pm'].mode()[0], inplace=True)
df['RainToday'].fillna(df['RainToday'].mode()[0], inplace=True)
df['RainTomorrow'].fillna(df['RainTomorrow'].mode()[0], inplace=True)
```

```python
#filling the null values in numerical variable with median
df = df.fillna(df.median())
```

```python
print('Jumlah Missing Value tiap column: ')
df.isnull().sum() #Check if any column has null values
```

Beberapa proses yang dilakukan di data cleaning:
1. Mendeklarasikan categorical variable dan numerical variable
2. Mengisi null value pada categorical variable dengan nilai mode-nya.
3. Mengisi null value pada numerical variable dengan nilai median-nya.
4. Delete Data column

# Modelling

**Label Encode**

```python
le = preprocessing.LabelEncoder()

df_model = df.copy()
for col in df_model.columns:
    df_model[col] = le.fit_transform(df_model[col])

df_model
```

**Determine X & Y**

```python
X = df_model.drop(['RainTomorrow'], axis=1)

y = df_model['RainTomorrow']
```

```python
X
```

**Splitting train and test data**

```python
# Split data 70:30
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=10)
```

```python
# Save user location for training and test set
train_id = np.array(X_train['Location'])
test_id = np.array(X_test['Location'])
```

**Feature Scaling**

```python
# Standard Scaler
ss_X = StandardScaler()
X_train = ss_X.fit_transform(X_train)
X_test = ss_X.transform(X_test)
```

**Function To Apply All Models**

```python
# Plotting ROC Curve
def plot_roc_cur(fper, tper):
    plt.plot(fper, tper, color='orange', label='ROC')
    plt.plot([0, 1], [0, 1], color='darkblue', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend()
    plt.show()


# Run given model
def run_model(model, X_train, y_train, X_test, y_test, verbose=True):
    t0 = time.time()
    if verbose == False:
        model.fit(X_train, y_train, verbose=0)
    else:
        model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    time_taken = time.time()-t0
    cm = confusion_matrix(y_test, y_pred)

    print(f'Training set score: {round(model.score(X_train, y_train) * 100, 2)}%')
    print('--'*40)
    print("Accuracy = {}%".format(round(accuracy*100, 2)))
    print('--'*40)
    print("ROC Area under Curve = {}".format(roc_auc))
    print('--'*40)
    print("Time taken = {}".format(time_taken))
    print('--'*40)
    print("Classification Report")
    print(classification_report(y_test,y_pred,digits=5))
    print('--'*40)
    print('Confusion Matrix\n\n', cm)
    print('--'*40)
    probs = model.predict_proba(X_test)
    probs = probs[:, 1]
    fper, tper, thresholds = roc_curve(y_test, probs)
    plot_roc_cur(fper, tper)

    plot_confusion_matrix(model, X_test, y_test,
                          cmap=plt.cm.Blues, normalize='all')
    return model, accuracy, time_taken
```

## Logistic Regression

```python
# Applying Logistic Regression on Data

LogR = LogisticRegression()
LogR, accuracy_dt,  tt_dt = run_model(LogR, X_train, y_train, X_test, y_test)
```
